# IMPORTING LIBRARIES

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# IMPORTING DATASET

```python
df=pd.read_csv('books.csv',error_bad_lines = False)
```

```python
df
```

```python
df.head()
```

# Analysis of Maximum Ratings Count,Text reviews count and Average Rating

```python
df[df['ratings_count']==np.max(df['ratings_count'])].set_index('title')
```

```python
df[df['text_reviews_count']==np.max(df['text_reviews_count'])].set_index('title')
```

```python
df[df['average_rating']==np.max(df['average_rating'])].set_index('title')
```

# Which are the top 10 books with most occurrences in the list?

```python
df['title'].value_counts()[:10]
```

```python
sns.set_context('poster')
plt.figure(figsize=(8,6))
books = df['title'].value_counts()[:10]
sns.barplot(x = books, y = books.index, palette='deep')
plt.title("Most Occurring Books")
plt.xlabel("Number of occurrences")
plt.ylabel("Books")
plt.show()
```

#  Average Ratings of first 10 Books

```python
df.set_index('title').average_rating[:10]
```

# What is the distribution of books for all languages?

```python
df.groupby('language_code')['title'].count()
```

```python
sns.set_context('paper')
plt.figure(figsize=(8,6))
ax = df.groupby('language_code')['title'].count().plot.bar()
plt.title('Language Code', fontsize=15)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
```

# Which are the top 10 most rated books?

```python
df.sort_values('ratings_count', ascending = False).head(10).set_index('title')
```

```python
most_rated = df.sort_values('ratings_count', ascending = False).head(10).set_index('title')
plt.figure(figsize=(10,8))
sns.barplot(most_rated['ratings_count'], most_rated.index, palette='rocket')
plt.title('Top 10 Rated Books', fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel('Ratings_count',fontsize=15)
plt.ylabel('Title',fontsize=15)
```

# What is the rating distribution for the books?

```python
df.average_rating.astype(float)
```

```python
plt.figure(figsize=(8,8))
rating= df.average_rating.astype(float)
sns.distplot(rating, bins=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel('Average_rating',fontsize=15)
plt.title('Rating Distribution',fontsize=15)
```

# Relationship between ratings and review counts

```python
plt.figure(figsize=(30,10))
df.dropna(0, inplace=True)
sns.set_context('paper')
ax =sns.jointplot(x="average_rating",y='text_reviews_count', kind='scatter',  data= df[['text_reviews_count', 'average_rating']])
ax.set_axis_labels("Average Rating", "Text Review Count")
plt.show()
```

# Relationship between Average Rating and Ratings Count

```python
sns.set_context('paper')
ax = sns.jointplot(x="average_rating", y="ratings_count", data = df, color = 'blueviolet')
ax.set_axis_labels("Average Rating", "Ratings Count")
```

# Which are the books with the highest reviews?

```python
df.sort_values('text_reviews_count', ascending = False).head(10).set_index('title')
```

```python
most_text = df.sort_values('text_reviews_count', ascending = False).head(10).set_index('title')
plt.figure(figsize=(15,10))
sns.set_context('poster')
ax = sns.barplot(most_text['text_reviews_count'], most_text.index, palette='magma')
plt.title('Books with highest reviews')
```

# Language code VS Average Rating Scatter Plot

```python
fig, ax = plt.subplots(figsize=(15,7))
ax.scatter(df['language_code'],df['average_rating'], color="red")
plt.title('Language code VS Average Rating Scatter Plot')
```
